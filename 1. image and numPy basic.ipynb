```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from PIL import Image ## python imaging library
```

```python
pic = Image.open('picS/dog.jpg') ## grabing the image
```

```python
## pic
type(pic) ## see file type
```

```python
pic_arr= np.asarray(pic) ## converting the pic as Numpy array
type(pic_arr)
```

```python
pic_arr.shape ## her output willbe (667, 1000, 3) ## here, 667,1000-> is resulation of pic(667,1000) and 3-> R,G,B 
```

```python
plt.imshow(pic_arr) ## show the image as pixel plot
```

```python
# NoW we gave 3 channel->  R G B
plt.imshow(pic_arr[:,:,0]) # show, only red channel image with all row collumn
```

```python
plt.imshow(pic_arr[:,:,0], cmap= 'gray') # red image in gray scale
```

```python
plt.imshow(pic_arr[:,:,1], cmap= 'gray')
```

```python
plt.imshow(pic_arr[:,:,2], cmap= 'gray')
```

```python
# Now, make a copy of array image and make the blue channel to zero
copy_image_array=pic_arr.copy() ## make a copy
copy_image_array[:,:,2] = 0 # blue value  zero
```

```python
plt.imshow(copy_image_array) # image will show without blue
```

```python
copy_image_array[:,:,1] = 0 # green value  zero
plt.imshow(copy_image_array) # only have red value
```
