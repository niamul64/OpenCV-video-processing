```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import cv2
```

```python
img=cv2.imread('picS/dog.jpg')
```

```python
type(img) # don't need to convert the image to arry, OpenCV directly converts it to
```

```python
img.shape
```

```python
plt.imshow(img) # img looks different because
                # MatplotLib -> RGB
                # openCV -> BGR
                # So we need to transform this
```

```python
fix_img = cv2.cvtColor( img , cv2.COLOR_BGR2RGB)
plt.imshow(fix_img)
```

```python
# we can read image directly as grayscale image
img_gray = cv2.imread('picS/dog.jpg', cv2.IMREAD_GRAYSCALE)
plt.imshow(img_gray, cmap='gray' )
```

```python
plt.imshow(fix_img)
```

```python
# we can resize the image
new_img= cv2.resize(fix_img, (1000,400))
```

```python
plt.imshow(new_img)
```

```python
# we can resize the image by ratio
w_ratio=0.95 #95%
h_ratio=0.2 # 20%
new_img2= cv2.resize(fix_img, (0,0), fix_img, w_ratio,h_ratio)
plt.imshow(new_img2)
```

```python
new_img2.shape
```

```python
#we can save the image
cv2.imwrite('picS/save.jpg',new_img2 ) #img will be saved
```

```python
# we can flip the img 
flipedi_img = cv2.flip(fix_img,0) # 0 for vertical flip
plt.imshow(flipedi_img)
```

```python
# we can flip the img 
flipedi_img2 = cv2.flip(fix_img,1) # 1 for horizental flip # -1 for vertical horizental
plt.imshow(flipedi_img2)
```
